```python
#since we have been able to derive the needed insight we now move to the machine learning aspects 
col=['ram', 'battery_power', 'px_height', 'px_width']
df[col]
```

```python
x=df[col].values
```

```python
x.shape
```

```python
y=df['price_range'].values
```

```python

y.shape
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test= train_test_split(x,y)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_test=sc.fit_transform(x_test)
x_train=sc.fit_transform(x_train)
```

```python
def model (x_train, y_train):
  from sklearn.linear_model import LogisticRegression
  logreg= LogisticRegression(random_state=0)
  logreg.fit(x_train, y_train)

  from sklearn.tree import DecisionTreeClassifier
  tree=DecisionTreeClassifier(criterion='entropy', random_state=0)
  tree.fit(x_train, y_train)

  from sklearn.ensemble import RandomForestClassifier
  forest=RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
  forest.fit(x_train, y_train)

  print('The accuracy of the logistic Regression Classifier is :',logreg.score(x_train, y_train))
  print('The accuracy of the Tree Classifier is : ', tree.score(x_train, y_train))
  print('The accuracy of the Forest classifier is: ', forest.score(x_train, y_train))

  return logreg, tree, forest
```

```python
models=model(x_train, y_train)
```

```python
for i in range(len(models)):
  print(models[i].predict(x_test))
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(models)):
  print(classification_report(y_test, models[i].predict(x_test)))
```
